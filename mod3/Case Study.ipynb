# Write a Python app that has the following classes:
1. A super class called Vehicle, which contains an attribute for vehicle type, such as car, truck, plane, boat, or a broomstick.


2. A class called Automobile which will inherit the attributes from Vehicle and also contain the following attributes:

year

make

model

doors (2 or 4)

roof (solid or sun roof).
  
3. Write an app that will accept user input for a car. The app will store "car" into the vehicle type in your Vehicle super class. 


4. The app will then ask the user for the year, make, model, doors, and type of roof and store thdata in the attributes above.


5. The app will then output the data in an easy-to-read and understandable format, such as this:

Vehicle type: car

Year: 2022

Make: Toyota

Model: Corolla

Number of doors: 4

Type of roof: sun roof

```python
class Vehicle:
    def __init__(self, vehicle_type: str) -> None:
        self.vehicle_type: str = vehicle_type

class Automobile(Vehicle):
    def __init__(self, vehicle_type: str, year: int, make: str, model: str, doors: int, roof: str) -> None:
        super().__init__(vehicle_type)
        self.year: int = year
        self.make: str = make
        self.model: str = model
        
        if doors == 2 or doors == 4: # (2 or 4)
            self.doors: int = doors
        else:
            raise ValueError("Doors must be 2 or 4")

        if roof.lower() == 'solid' or roof.lower() == 'sunroof':
            self.roof: str = roof.lower()       # (solid or sunroof)
        else:
            raise ValueError("Roof must be solid or sunroof")
year: int = int(input('input year: '))
make: str = input('input make of car: ')
model: str = input('input model of car: ')
doors: int = int(input('input doors (2 or 4): '))
roof: str = input('input type of roof (solid or sunroof): ')
car = Automobile('car', year, make, model, doors, roof)
print(f"Vehicle type: {car.vehicle_type}\n Year: {car.year}\n Make: {car.make}\n Model: {car.model}\n Number of doors: {car.doors}\n Roof type: {car.roof}")
```
