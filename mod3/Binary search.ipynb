# binary search 
First code is a number that exists within the index and the second is a number that is not found in the index

```python
arr: list[int] = [1, 2, 3, 4, 5, 6, 7, 8, 9]
k = 4

is_found: bool = False
for i in range(len(arr)):
    if k == arr[i]:
        is_found = True
        print(f"number {k} found at index {i}")
        break
if is_found == False:
    print(f"number {k} not found in the array")
```
