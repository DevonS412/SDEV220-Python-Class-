# sort the numbers in the list in ascending order 

```python
def sort(arr):
    if len(arr) <=1:
        return arr
    pivot = arr[0]
    left = [x for x in arr[1:] if x < pivot]
    right = [x for x in arr[1:] if x >= pivot]

    return sort(left) + [pivot] + sort(right)

arr = [1, 0, 2, 0, 1, 2, 1, 0, 2]
sorted_arr = sort(arr)
print(sorted_arr)
```
